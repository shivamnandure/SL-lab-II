# Sentiment Analysis using KNN on Twitter Dataset

## Aim
To perform sentiment analysis on a Twitter dataset using the K-Nearest Neighbors (KNN) algorithm.

This notebook includes:
- Data loading
- Data preprocessing
- TF-IDF vectorization
- KNN model training
- Model evaluation
- Visualization graphs and plots
- Custom tweet prediction


```
# =========================================
# Step 1: Import Libraries
# =========================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print("Libraries Imported Successfully!")
```

```
# =========================================
# Step 2: Load Dataset
# =========================================

# Upload your dataset in Google Colab or keep it in same folder in Jupyter

df = pd.read_csv('tweets _dataset.csv', encoding='latin1')

print("Dataset Loaded Successfully!")

# Display first 5 rows
df.head()
```

```
# =========================================
# Step 3: Dataset Information
# =========================================

print(df.info())

print("\nMissing Values:\n")
print(df.isnull().sum())
```

```
# =========================================
# Step 4: Sentiment Distribution Visualization
# =========================================

plt.figure(figsize=(6,5))

sns.countplot(x='sentiment', data=df)

plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')

plt.show()
```

```
# =========================================
# Step 5: Select Features and Labels
# =========================================

X = df['text']
y = df['sentiment']

print("Features and Labels Selected!")
```

```
# =========================================
# Step 6: Convert Text into Numerical Form
# =========================================

vectorizer = TfidfVectorizer(stop_words='english')

X_vectorized = vectorizer.fit_transform(X)

print("Text Vectorization Completed!")
print("Shape of Vectorized Data:", X_vectorized.shape)
```

```
# =========================================
# Step 7: Split Dataset
# =========================================

X_train, X_test, y_train, y_test = train_test_split(
    X_vectorized,
    y,
    test_size=0.2,
    random_state=42
)

print("Training Data Shape:", X_train.shape)
print("Testing Data Shape:", X_test.shape)
```

```
# =========================================
# Step 8: Train KNN Model
# =========================================

knn = KNeighborsClassifier(n_neighbors=5)

knn.fit(X_train, y_train)

print("KNN Model Trained Successfully!")
```

```
# =========================================
# Step 9: Make Predictions
# =========================================

y_pred = knn.predict(X_test)

print("Predictions Completed!")
```

```
# =========================================
# Step 10: Model Accuracy
# =========================================

accuracy = accuracy_score(y_test, y_pred)

print("Model Accuracy:", accuracy)
```

```
# =========================================
# Step 11: Classification Report
# =========================================

print(classification_report(y_test, y_pred))
```

```
# =========================================
# Step 12: Confusion Matrix
# =========================================

cm = confusion_matrix(y_test, y_pred)

print(cm)
```

```
# =========================================
# Step 13: Confusion Matrix Heatmap
# =========================================

plt.figure(figsize=(6,5))

sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')

plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')

plt.show()
```

```
# =========================================
# Step 14: Accuracy Bar Plot
# =========================================

plt.figure(figsize=(5,5))

plt.bar(['Accuracy'], [accuracy])

plt.title('Model Accuracy')

plt.ylim(0,1)

plt.show()
```

```
# =========================================
# Step 15: Custom Tweet Prediction
# =========================================

sample_tweet = ["I really love this product and it is amazing!"]

sample_vector = vectorizer.transform(sample_tweet)

prediction = knn.predict(sample_vector)

print("Tweet:", sample_tweet[0])
print("Predicted Sentiment:", prediction[0])
```
