```

# ============================================================
# Assignment 5 : K-Means Clustering
# ============================================================

# STEP 1 : Import Libraries

import numpy as np
import matplotlib.pyplot as plt

# STEP 2 : Define Dataset

P1 = [0.1, 0.6]
P2 = [0.15, 0.71]
P3 = [0.08, 0.9]
P4 = [0.16, 0.85]
P5 = [0.2, 0.3]
P6 = [0.25, 0.5]
P7 = [0.24, 0.1]
P8 = [0.3, 0.2]

points = {
    "P1": np.array(P1),
    "P2": np.array(P2),
    "P3": np.array(P3),
    "P4": np.array(P4),
    "P5": np.array(P5),
    "P6": np.array(P6),
    "P7": np.array(P7),
    "P8": np.array(P8)
}

print("Dataset Loaded Successfully")

# STEP 3 : Initialize Centroids

m1 = np.array(P1)
m2 = np.array(P8)

print("\nInitial Centroids")
print("m1 =", m1)
print("m2 =", m2)

# STEP 4 : Empty Clusters

C1 = []
C2 = []

# STEP 5 : Distance Calculation

print("\n----- Distance Calculation -----\n")

for name, point in points.items():

    d1 = np.linalg.norm(point - m1)
    d2 = np.linalg.norm(point - m2)

    print(f"{name}")
    print("Distance from m1 =", round(d1, 4))
    print("Distance from m2 =", round(d2, 4))

    if d1 < d2:
        C1.append((name, point))
        print("Assigned to Cluster C1\n")
    else:
        C2.append((name, point))
        print("Assigned to Cluster C2\n")

# STEP 6 : Display Clusters

print("Cluster C1 Points:")
for point in C1:
    print(point[0])

print("\nCluster C2 Points:")
for point in C2:
    print(point[0])

# STEP 7 : Update Centroids

m1_new = np.mean([point[1] for point in C1], axis=0)
m2_new = np.mean([point[1] for point in C2], axis=0)

print("\nUpdated Centroids")
print("Updated m1 =", m1_new)
print("Updated m2 =", m2_new)

# STEP 8 : Final Answers

if any(point[0] == "P6" for point in C1):
    p6_cluster = "C1"
else:
    p6_cluster = "C2"

print("\n================ FINAL ANSWERS ================")

print("1] P6 belongs to :", p6_cluster)
print("2] Population around m2 :", len(C2))
print("3] Updated m1 =", m1_new)
print("   Updated m2 =", m2_new)

# STEP 9 : Visualization

plt.figure(figsize=(10,7))

# Cluster C1
for name, point in C1:
    plt.scatter(point[0], point[1], s=100)
    plt.text(point[0]+0.005, point[1]+0.005, name)

# Cluster C2
for name, point in C2:
    plt.scatter(point[0], point[1], s=100)
    plt.text(point[0]+0.005, point[1]+0.005, name)

# Initial Centroids
plt.scatter(m1[0], m1[1], marker='X', s=300, label='Initial m1')
plt.scatter(m2[0], m2[1], marker='X', s=300, label='Initial m2')

# Updated Centroids
plt.scatter(m1_new[0], m1_new[1], marker='D', s=250, label='Updated m1')
plt.scatter(m2_new[0], m2_new[1], marker='D', s=250, label='Updated m2')

plt.title("K-Means Clustering Visualization")
plt.xlabel("X Coordinate")
plt.ylabel("Y Coordinate")
plt.grid(True)
plt.legend()

plt.show()
```
