# Efficient Back Propagation Neural Network for XOR Function

```
import numpy as np
import matplotlib.pyplot as plt
```

## XOR Dataset

```
# XOR dataset
X = np.array([[0,0],[0,1],[1,0],[1,1]])
Y = np.array([[0],[1],[1],[0]])

print("Input:\n", X)
print("\nExpected Output:\n", Y)
```

## Sigmoid Function

```
# Sigmoid function
sigmoid = lambda x: 1/(1+np.exp(-x))
derivative = lambda x: x*(1-x)
```

## Initialize Weights

```
# Random weights
np.random.seed(1)
w1 = np.random.rand(2,2)
w2 = np.random.rand(2,1)

print("Weights between Input and Hidden Layer:\n", w1)
print("\nWeights between Hidden and Output Layer:\n", w2)
```

## Training Using Backpropagation

```
errors = []

# Training
for i in range(10000):

    # Forward propagation
    h = sigmoid(np.dot(X,w1))
    o = sigmoid(np.dot(h,w2))

    # Error
    e = Y-o
    errors.append(np.mean(np.abs(e)))

    # Backpropagation
    d_o = e * derivative(o)
    d_h = d_o.dot(w2.T) * derivative(h)

    # Weight update
    w2 += h.T.dot(d_o) * 0.1
    w1 += X.T.dot(d_h) * 0.1
```

## Final Output

```
print("Predicted Output:\n")
print(o)
```

## Error Graph

```
plt.plot(errors)

plt.title("Error Reduction")
plt.xlabel("Epochs")
plt.ylabel("Error")

plt.grid()
plt.show()
```
