# Adaptive Resonance Theory (ART1)

Adaptive Resonance Theory (ART1) is a neural network model used for clustering binary patterns.  
It groups similar input patterns into clusters using a vigilance parameter.

- Higher vigilance → More clusters
- Lower vigilance → Fewer clusters


```

# ============================================================
# ADAPTIVE RESONANCE THEORY (ART1) - GOOGLE COLAB / JUPYTER
# ============================================================

# Import Libraries
import numpy as np
import matplotlib.pyplot as plt

# ------------------------------------------------------------
# STEP 1: Define Input Patterns
# ------------------------------------------------------------

patterns = np.array([
    [1, 1, 0, 0],
    [1, 0, 0, 0],
    [0, 0, 1, 1],
    [0, 0, 1, 0]
])

print("Input Patterns:\n")
print(patterns)

# ------------------------------------------------------------
# STEP 2: Set Vigilance Parameter
# ------------------------------------------------------------

vigilance = 0.6

print("\nVigilance Parameter =", vigilance)

# ------------------------------------------------------------
# STEP 3: Initialize Clusters
# ------------------------------------------------------------

clusters = []
cluster_labels = []

# ------------------------------------------------------------
# STEP 4: Similarity Function
# ------------------------------------------------------------

def similarity(p1, p2):

    match = np.sum(np.logical_and(p1, p2))
    total = np.sum(p1)

    if total == 0:
        return 0

    return match / total

# ------------------------------------------------------------
# STEP 5: ART1 Clustering Process
# ------------------------------------------------------------

print("\n--- ART1 Clustering Process ---\n")

for pattern in patterns:

    assigned = False

    # Compare with existing clusters
    for i, cluster in enumerate(clusters):

        sim = similarity(pattern, cluster)

        print(f"Comparing {pattern} with Cluster {i+1}")
        print("Similarity =", sim)

        # Check vigilance condition
        if sim >= vigilance:

            print(f"Pattern {pattern} assigned to Cluster {i+1}\n")

            # Update cluster prototype
            clusters[i] = np.logical_and(cluster, pattern).astype(int)

            cluster_labels.append(i + 1)

            assigned = True
            break

    # Create new cluster if not assigned
    if not assigned:

        clusters.append(pattern.copy())

        cluster_labels.append(len(clusters))

        print(f"Pattern {pattern} creates NEW Cluster {len(clusters)}\n")

# ------------------------------------------------------------
# STEP 6: Display Final Clusters
# ------------------------------------------------------------

print("\nFinal Clusters:\n")

for i, cluster in enumerate(clusters):

    print(f"Cluster {i+1} : {cluster}")

# ------------------------------------------------------------
# STEP 7: Visualization
# ------------------------------------------------------------

x = np.arange(1, len(patterns) + 1)

plt.figure(figsize=(8,5))

plt.bar(x, cluster_labels)

plt.xlabel("Pattern Number")
plt.ylabel("Cluster Number")
plt.title("ART1 Clustering Visualization")

plt.xticks(x)

plt.grid(axis='y')

plt.show()
```
