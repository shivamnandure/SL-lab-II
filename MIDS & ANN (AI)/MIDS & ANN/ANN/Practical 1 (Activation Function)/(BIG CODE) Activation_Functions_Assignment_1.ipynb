## Activation Functions in Neural Networks

```

# Import Libraries
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid Function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# ReLU Function
def relu(x):
    return np.maximum(0, x)

# Tanh Function
def tanh(x):
    return np.tanh(x)

# Leaky ReLU Function
def leaky_relu(x):
    return np.where(x > 0, x, 0.01 * x)

# Softmax Function
def softmax(x):
    exp_values = np.exp(x - np.max(x))
    return exp_values / np.sum(exp_values)

# Take User Input
value = float(input("Enter a number: "))

# Display Outputs
print("\n========== Activation Function Outputs ==========\n")

print("Input Value :", value)

print("\nSigmoid Output      :", sigmoid(value))
print("ReLU Output         :", relu(value))
print("Tanh Output         :", tanh(value))
print("Leaky ReLU Output   :", leaky_relu(value))

# Softmax Example
sample_array = np.array([1.0, 2.0, 3.0])

print("\nSoftmax Input  :", sample_array)
print("Softmax Output :", softmax(sample_array))

# Generate Values for Graph
x = np.linspace(-10, 10, 200)

# Plot Activation Functions
plt.figure(figsize=(12, 7))

plt.plot(x, sigmoid(x), label='Sigmoid')
plt.plot(x, relu(x), label='ReLU')
plt.plot(x, tanh(x), label='Tanh')
plt.plot(x, leaky_relu(x), label='Leaky ReLU')

# Graph Formatting
plt.title("Activation Functions in Neural Networks", fontsize=16)

plt.xlabel("Input Values", fontsize=12)
plt.ylabel("Output Values", fontsize=12)

plt.grid(True)
plt.legend()

# Show Graph
plt.show()
```
