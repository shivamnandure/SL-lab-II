# Perceptron Learning Law with Decision Region
This notebook demonstrates the perceptron learning algorithm with graphical visualization using Python.

```
import numpy as np
import matplotlib.pyplot as plt
```

## Define Training Data

```
# Input Data
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

# Target Output
y = np.array([0, 0, 0, 1])

print("Input Data:")
print(X)

print("\nTarget Output:")
print(y)
```

## Initialize Perceptron Parameters

```
# Initialize weights and bias
weights = np.zeros(2)
bias = 0

# Learning rate
learning_rate = 0.1

# Number of epochs
epochs = 10

print("Initial Weights:", weights)
print("Initial Bias:", bias)
```

## Train the Perceptron

```
for epoch in range(epochs):

    print(f"\nEpoch {epoch+1}")

    for i in range(len(X)):

        # Calculate net input
        net_input = np.dot(X[i], weights) + bias

        # Activation function
        y_pred = 1 if net_input >= 0 else 0

        # Calculate error
        error = y[i] - y_pred

        # Update weights and bias
        weights = weights + learning_rate * error * X[i]
        bias = bias + learning_rate * error

        print(f"Input: {X[i]}  Target: {y[i]}  Predicted: {y_pred}")

print("\nTraining Complete")
```

## Display Final Weights

```
print("Final Weights:", weights)
print("Final Bias:", bias)
```

## Plot Decision Region

```
# Define graph limits
x_min, x_max = -0.5, 1.5
y_min, y_max = -0.5, 1.5

# Create meshgrid
xx, yy = np.meshgrid(
    np.arange(x_min, x_max, 0.01),
    np.arange(y_min, y_max, 0.01)
)

# Predict output for each point
Z = np.dot(np.c_[xx.ravel(), yy.ravel()], weights) + bias

# Apply activation function
Z = np.where(Z >= 0, 1, 0)

# Reshape for contour plot
Z = Z.reshape(xx.shape)

# Plot decision region
plt.contourf(xx, yy, Z, alpha=0.3, cmap='coolwarm')

# Plot training points
for i in range(len(X)):

    if y[i] == 0:
        plt.scatter(X[i][0], X[i][1],
                    color='red',
                    marker='o',
                    s=100,
                    label='Class 0' if i == 0 else "")

    else:
        plt.scatter(X[i][0], X[i][1],
                    color='blue',
                    marker='x',
                    s=100,
                    label='Class 1')

# Labels and title
plt.title("Perceptron Learning Law with Decision Region")
plt.xlabel("Input X1")
plt.ylabel("Input X2")

plt.legend()
plt.grid(True)

plt.show()
```
